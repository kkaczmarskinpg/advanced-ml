## Podstawowe podsumowanie zbioru danych

W tej sekcji przedstawiamy podstawowe informacje o zbiorze: rozmiar, typy danych, statystyki opisowe oraz listę cech.

```python
import pandas as pd

# Wczytanie danych z lokalnego pliku CSV
csv_path = "./data/loan_data.csv"
df = pd.read_csv(csv_path)

print("Rozmiar zbioru (wiersze, kolumny):", df.shape)
print("\nInformacje o kolumnach:")
df.info()

print("\nStatystyki opisowe dla kolumn numerycznych:")
display(df.describe())

print("\nStatystyki opisowe dla wszystkich kolumn:")
display(df.describe(include="all"))

print("\nTypy danych:")
display(df.dtypes.to_frame("dtype"))
```

### Lista cech i ich znaczenie

- `person_age` - wiek osoby
- `person_gender` - płeć
- `person_education` - poziom wykształcenia
- `person_income` - roczny dochód
- `person_emp_exp` - doświadczenie zawodowe (lata)
- `person_home_ownership` - status mieszkaniowy (np. RENT, OWN, MORTGAGE)
- `loan_amnt` - kwota pożyczki
- `loan_intent` - cel pożyczki
- `loan_int_rate` - oprocentowanie pożyczki
- `loan_percent_income` - udział kwoty pożyczki w dochodzie
- `cb_person_cred_hist_length` - długość historii kredytowej
- `credit_score` - scoring kredytowy
- `previous_loan_defaults_on_file` - informacja o wcześniejszych problemach ze spłatą
- `loan_status` - zmienna zależna (etykieta klasyfikacji: 0/1)

Zbiór danych nadaje się do klasyfikacji binarnej. Zmienną zależną jest `loan_status`, a pozostałe cechy opisują profil klienta, parametry pożyczki oraz historię kredytową.
