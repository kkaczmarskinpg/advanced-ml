## Imputacja danych błędnych i brakujących

 > Założenie: kolumna docelowa `loan_status` nie może mieć braków i nie imputujemy jej.

W tym punkcie realizujemy 4 podejścia:
1. Prosta imputacja A: mediana (numeryczne) + moda (kategoryczne).
2. Prosta imputacja B: średnia (numeryczne) + moda (kategoryczne).
3. Imputacja modelowa: `KNNImputer` dla cech numerycznych + moda dla kategorycznych.
4. Imputacja AutoML: predykcja brakujących wartości w każdej kolumnie z użyciem AutoGluon.

```python
import pandas as pd
import numpy as np

from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer

# Wczytanie danych z brakami wygenerowanymi wcześniej
csv_missing_path = "./data/loan_data_with_missing.csv"
df_missing = pd.read_csv(csv_missing_path)

# Kolumna klasyfikacyjna - nie imputujemy jej
target_col = "loan_status"
X = df_missing.drop(columns=[target_col]).copy()
y = df_missing[target_col].copy()

num_cols = X.select_dtypes(include=["number"]).columns.tolist()
cat_cols = X.select_dtypes(exclude=["number"]).columns.tolist()

print("Rozmiar X:", X.shape)
print("Braki w X (łącznie):", int(X.isna().sum().sum()))
print("Braki w y:", int(y.isna().sum()))
print("Kolumny numeryczne:", len(num_cols), "| kategoryczne:", len(cat_cols))

def raport_brakow(df, nazwa):
    print(f"\n[{nazwa}] Braki łącznie: {int(df.isna().sum().sum())}")
    display(df.isna().mean().sort_values(ascending=False).head(10).to_frame("missing_ratio"))
```

```python
# --- 1) Prosta imputacja A: mediana + moda ---
ct_simple_a = ColumnTransformer(
    transformers=[
        ("num", SimpleImputer(strategy="median"), num_cols),
        ("cat", SimpleImputer(strategy="most_frequent"), cat_cols),
    ],
    remainder="drop",
    verbose_feature_names_out=False,
).set_output(transform="pandas")

X_simple_a = ct_simple_a.fit_transform(X)
raport_brakow(X_simple_a, "Prosta A")

# --- 2) Prosta imputacja B: średnia + moda ---
ct_simple_b = ColumnTransformer(
    transformers=[
        ("num", SimpleImputer(strategy="mean"), num_cols),
        ("cat", SimpleImputer(strategy="most_frequent"), cat_cols),
    ],
    remainder="drop",
    verbose_feature_names_out=False,
).set_output(transform="pandas")

X_simple_b = ct_simple_b.fit_transform(X)
raport_brakow(X_simple_b, "Prosta B")

# --- 3) Imputacja modelowa: KNNImputer + moda ---
from sklearn.impute import KNNImputer

# Numeryczne: imputacja modelowa oparta o najblizszych sasiadow
knn_imp = KNNImputer(n_neighbors=5, weights="uniform")
X_num_knn = pd.DataFrame(knn_imp.fit_transform(X[num_cols]), columns=num_cols, index=X.index)

# Kategoryczne: prosta moda
cat_imp = SimpleImputer(strategy="most_frequent")
X_cat_imp = pd.DataFrame(cat_imp.fit_transform(X[cat_cols]), columns=cat_cols, index=X.index)

X_model_knn = pd.concat([X_num_knn, X_cat_imp], axis=1)[X.columns]
raport_brakow(X_model_knn, "Modelowa (KNNImputer)")

# Gotowe zestawy do dalszego modelowania klasyfikacji:
# X_simple_a, X_simple_b, X_model_knn oraz y (bez brakow)
```

```python
# --- 4) Imputacja AutoML: AutoGluon dla każdej kolumny z brakami ---
# Uwaga: to podejście jest kosztowne obliczeniowo. Trenujemy osobny model dla każdej kolumny z brakami.

df_automl = X.copy()
missing_counts = df_automl.isna().sum().sort_values(ascending=False)
cols_with_missing = missing_counts[missing_counts > 0].index.tolist()

if len(cols_with_missing) == 0:
    print("Brak brakujących wartości w cechach - AutoML nie jest potrzebny.")
else:
    try:
        from autogluon.tabular import TabularPredictor

        successful_cols = []
        failed_cols = []

        for col_to_impute in cols_with_missing:
            print(f"\nImputacja AutoML dla kolumny: {col_to_impute}")

            known_mask = df_automl[col_to_impute].notna()
            train_df = df_automl.loc[known_mask].copy()
            pred_df = df_automl.loc[~known_mask].copy()

            if len(pred_df) == 0:
                print("  Brak braków do uzupełnienia - pomijam.")
                continue
            if len(train_df) < 50:
                print("  Za mało obserwacji uczących - pomijam.")
                failed_cols.append(col_to_impute)
                continue

            # Uzupełnienie cech wejściowych, aby AutoML mogło trenować
            for c in train_df.columns:
                if c == col_to_impute:
                    continue
                if pd.api.types.is_numeric_dtype(train_df[c]):
                    med = train_df[c].median()
                    train_df[c] = train_df[c].fillna(med)
                    pred_df[c] = pred_df[c].fillna(med)
                else:
                    mode_vals = train_df[c].mode(dropna=True)
                    fill_val = mode_vals.iloc[0] if len(mode_vals) > 0 else "BRAK"
                    train_df[c] = train_df[c].fillna(fill_val)
                    pred_df[c] = pred_df[c].fillna(fill_val)

            # Typ zadania dla imputowanej kolumny
            if pd.api.types.is_numeric_dtype(train_df[col_to_impute]):
                problem_type = "regression"
            else:
                problem_type = "multiclass"
                train_df[col_to_impute] = train_df[col_to_impute].astype(str)

            try:
                predictor = TabularPredictor(
                    label=col_to_impute,
                    problem_type=problem_type
                ).fit(
                    train_data=train_df,
                    time_limit=45,
                    presets="medium_quality",
                    verbosity=0
                )

                preds = predictor.predict(pred_df.drop(columns=[col_to_impute]))
                if problem_type == "multiclass":
                    preds = preds.astype(object)

                df_automl.loc[~known_mask, col_to_impute] = preds.values
                successful_cols.append(col_to_impute)
                print(f"  OK - uzupełniono {len(pred_df)} braków")

            except Exception as e:
                failed_cols.append(col_to_impute)
                print(f"  Błąd dla kolumny {col_to_impute}: {e}")

        print("\nPodsumowanie AutoML:")
        print("Udało się imputować kolumny:", successful_cols)
        print("Nie udało się imputować kolumn:", failed_cols)
        raport_brakow(df_automl, "AutoML (wszystkie kolumny)")

    except ImportError:
        print("AutoGluon nie jest zainstalowany. Zainstaluj: pip install autogluon.tabular")
        print("Po instalacji uruchom ponownie tę komórkę.")
```

```python
# --- Skalowanie i zapis 8 zbiorow (4 imputacje x 2 skalowania) ---
from pathlib import Path
from sklearn.preprocessing import MinMaxScaler, StandardScaler

output_dir = Path("./data")
output_dir.mkdir(parents=True, exist_ok=True)

# Wszystkie 4 metody imputacji
imputed_datasets = {
    "simple_a": X_simple_a.copy(),
    "simple_b": X_simple_b.copy(),
    "model_knn": X_model_knn.copy(),
    "automl": df_automl.copy(),
}

def scale_numeric_columns(features_df: pd.DataFrame, scaler):
    """Skalowanie tylko kolumn numerycznych, kategorie pozostaja bez zmian."""
    scaled_df = features_df.copy()
    local_num_cols = scaled_df.select_dtypes(include=["number"]).columns.tolist()
    if local_num_cols:
        scaled_df[local_num_cols] = scaler.fit_transform(scaled_df[local_num_cols])
    return scaled_df

scalers = {
    "minmax": MinMaxScaler(),
    "standard": StandardScaler(),
}

saved_files = []

for imputation_name, features_df in imputed_datasets.items():
    for scaler_name, scaler_obj in scalers.items():
        X_scaled = scale_numeric_columns(features_df, scaler_obj)

        full_df = X_scaled.copy()
        full_df[target_col] = y.values

        out_path = output_dir / f"loan_data_{imputation_name}_{scaler_name}.csv"
        full_df.to_csv(out_path, index=False)

        saved_files.append(out_path)
        print(f"Zapisano: {out_path} | kształt: {full_df.shape}")

print(f"\nLiczba zapisanych zbiorow: {len(saved_files)}")
```
